```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import model_selection
from sklearn.model_selection import train_test_split, learning_curve


# Load the data
df = pd.read_csv('../data/new_titanic_features.csv')


# Create Features and Labels
features = df[['Male', 'Family',
               'Pclass2_one', 'Pclass2_two', 'Pclass2_three',
               'Embarked_C', 'Embarked_Q', 'Embarked_S',
               'Age2', 'Fare3_Fare11to50', 'Fare3_Fare51+', 'Fare3_Fare<=10']]


#print(features.info)

features['Age2'] = 1000*features['Age2']


# Train test split
x_train, x_test, y_train, y_test = train_test_split(features,
                                                    label,
                                                    test_size=.2,
                                                    random_state=0)


# Create instance of Logistic Regression classifier
model = LogisticRegression(C=1, random_state=0)
lrc = model.fit(x_train, y_train)
pred_train = lrc.predict(x_train)


print('Training Accuracy:', metrics.accuracy_score(pred_train, y_train))

pred_test = lrc.predict(x_test)

print('Test Accuracy:', metrics.accuracy_score(y_test, pred_test))
print('Confusion Matrix:')
print(metrics.confusion_matrix(y_test, pred_test))

print('Classification Report:')
print(metrics.classification_report(y_test, pred_test))


# Get feature importance
coeffs = pd.DataFrame(list(zip(features.columns, lrc.coef_.ravel())),
                      columns=['feature', 'coefficients'])
print("Feature importances:", coeffs)


# Cross-validation
cv = model_selection.ShuffleSplit(n_splits=5, test_size=.4, random_state=0)
scores = model_selection.cross_val_score(model, features, label, cv=cv)
print(scores)
print('Crossval score: %0.2f +/- %0.2f ' % (scores.mean(), scores.std()))


# Learning Curve
tsz = np.linspace(0.1, 1, 10)
train_sizes, train_scores, test_scores = learning_curve(lrc, features, label,
                                                        train_sizes=tsz)

fig = plt.figure()
plt.plot(train_sizes, train_scores.mean(axis=1), 'ro-', label="Train Scores")
plt.plot(train_sizes, test_scores.mean(axis=1), 'go-', label="Test Scores")
plt.title('Learning Curve: Logistic Regression')
plt.ylim((0.5, 1.0))
plt.legend()
plt.draw()
plt.show()


# Exercises
#
# 1) Try rescaling the age feature with the standard scaler.
#    Does that change the result?
#    http://scikit-learn.org/stable/modules/preprocessing.html

'''
Training Accuracy: 0.821378340366
Test Accuracy: 0.719101123596
Confusion Matrix:
[[86 19]
 [31 42]]
Classification Report:
             precision    recall  f1-score   support

          0       0.74      0.82      0.77       105
          1       0.69      0.58      0.63        73

avg / total       0.72      0.72      0.71       178

Feature importances:              feature  coefficients
0               Male     -2.594783
1             Family     -0.056240
2        Pclass2_one      1.309481
3        Pclass2_two      0.614350
4      Pclass2_three     -0.606129
5         Embarked_C      0.536608
6         Embarked_Q      0.885574
7         Embarked_S     -0.104481
8               Age2     -0.030736
9   Fare3_Fare11to50      0.342374
10     Fare3_Fare51+      0.719065
11    Fare3_Fare<=10      0.256262
[ 0.76123596  0.78370787  0.78651685  0.78932584  0.82303371]
Crossval score: 0.79 +/- 0.02 
'''


'''
1000*age

Training Accuracy: 0.783403656821
Test Accuracy: 0.73595505618
Confusion Matrix:
[[97  8]
 [39 34]]
Classification Report:
             precision    recall  f1-score   support

          0       0.71      0.92      0.80       105
          1       0.81      0.47      0.59        73

avg / total       0.75      0.74      0.72       178

Feature importances:              feature  coefficients
0               Male -9.872959e-01
1             Family  3.338556e-01
2        Pclass2_one  4.278295e-01
3        Pclass2_two  1.568129e-01
4      Pclass2_three -5.301183e-01
5         Embarked_C  2.325014e-01
6         Embarked_Q  6.609608e-02
7         Embarked_S -2.440733e-01
8               Age2  6.366182e-07
9   Fare3_Fare11to50  1.049949e-01
10     Fare3_Fare51+  3.906577e-01
11    Fare3_Fare<=10 -4.411284e-01
[ 0.63483146  0.74157303  0.8005618   0.7752809   0.81741573]
Crossval score: 0.75 +/- 0.06 
'''

# 2) Experiment with another classifier for example
#    DecisionTreeClassifier or SVC
#    you can find classifiers here:
#    http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
#    Which ones are impacted by the age rescale? Why?

from sklearn import tree
dt = tree.DecisionTreeClassifier(min_samples_split=20, random_state=99)
dt.fit(x_train, y_train)
pred_train = dt.predict(x_test)
print("score on decision tree",dt.score(x_test, y_test, sample_weight=None))
'''
score on decision tree 0.73595505618

age=1000*age
score on decision tree 0.73595505618

There is no change in the score


'''
```
